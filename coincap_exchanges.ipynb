```python
import requests
import pandas as pd
url = 'https://api.coincap.io/v2/exchanges'
response = requests.get(url)
if response.ok:
    df = pd.read_json(response.text)
    print(df)
else:
    print("Failed to retrieve data. Status code:", response.status_code)
```

```python
import json
df_data = pd.json_normalize(df['data'])
df_data
```

```python
df_data.to_csv('exchanges.csv', index=False)
```

```python
df.to_json('exchanges1.json' )
df_data.to_json('exchanges2.json' )
```

```python
df_data.to_xml('exchanges.xml', index=False)
```

```python
df_data.to_parquet('exchanges.parquet', index=False)
```
